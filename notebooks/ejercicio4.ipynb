```python
from utils import manejo_archivos
from utils.constantes import DATASETS_PATH_RAW

# Metodo para obtener config del datasets
from utils.config_archivo import get_core_info

from ejercicios import ejercicio4
```

```python
# Construir ruta al archivos
datasets_paths = [
    DATASETS_PATH_RAW / 'iadiza',
    DATASETS_PATH_RAW / 'inaturalist',
    DATASETS_PATH_RAW / 'xeno_canto'
]
configs = []

# Cargar Configuracion

for i in range(len(datasets_paths)):
    config, core = get_core_info(datasets_paths[i])
    datasets_paths[i] = datasets_paths[i] / core
    configs.append(config)

 
```

```python
# -------------------------------inciso a y b--------------------------------
"""
Definir una estructura (tupla, lista, diccionario, clase) en Python que represente un registro 
para cada uno de los datasets. No considerar el ID. 

Escribir una función que: 
● reciba la lista de columnas del dataset 
● genere un registro vacío, donde todos los valores estén inicializados (por ejemplo, 
con None, cadena vacía, 0, etc.) 
Esto facilitará la creación de nuevos registros válidos.  No considerar el ID. """


diccionario_registros = []

for i in range(len(configs)):
    #generamos el registro vacio usando el header para las claves del diccionario
    header = manejo_archivos.get_header(datasets_paths[i], **configs[i])
    registro_vacio = ejercicio4.generar_registro_vacio(header)
    datasets = datasets_paths[i].parent.name
    diccionario_registros.append({datasets: registro_vacio})

print(header)
```

```python
#--------------------------------inciso c y d--------------------------------
"""Escribir una función que valide un registro antes de insertarlo. Debe reutilizar todas las 
funciones posibles ya desarrolladas para este fin.  No considerar el ID.

Escribir una función que reciba la estructura definida en el inciso 4.A y retorne una 
lista/diccionario listo para ser añadido al csv que contiene el dataset. 
Deberá considerar añadir el/los IDs a la información deduciendo cómo se construyen de 
acuerdo a cada dataset (autoincremental, random, etc)."""

for i in range(len(configs)):
    #completamos el registro con la informacion y validamos el registro
    ejercicio4.completar_registro(diccionario_registros[i], **configs[i])
    validar_registro = ejercicio4.validar_registro(diccionario_registros[i], **configs[i])
    
    #obtenemos el nombre del dataset , generamos el ID para el registro y lo agregamos al diccionario
    datasets = datasets_paths[i].parent.name
    archivo = manejo_archivos.get_archive(datasets_paths[i], **configs[i])
    generar_id = ejercicio4.generar_id(archivo,datasets)
    diccionario_registros[i][datasets]['id'] = generar_id

    
    print(f"Registro con ID generado para el dataset {datasets}: {diccionario_registros[i]}")
```
