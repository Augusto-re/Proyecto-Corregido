```python
from utils import manejo_archivos
from utils.constantes import DATASETS_PATH_RAW

# Metodo para obtener config del datasets
from utils.config_archivo import get_core_info

from ejercicios import ejercicio4
```

```python
# Construir ruta al archivos
datasets_paths = [
    DATASETS_PATH_RAW / 'iadiza',
    DATASETS_PATH_RAW / 'inaturalist',
    DATASETS_PATH_RAW / 'xeno_canto'
]
configs = []

# Cargar Configuracion

for i in range(len(datasets_paths)):

    nombre_carpeta = datasets_paths[i].name 
    
    config, core = get_core_info(datasets_paths[i])

    nombre_final = f"{nombre_carpeta}_{core}"

    config['core'] = nombre_final
    datasets_paths[i] = datasets_paths[i] / core
    configs.append(config)
```

```python

#--------------------------------Ejercicio 4 COMPLETO--------------------------------
print("INSERTAR REGISTRO EN EL DATASET PROCESADO")
print("---------------------------------------------")
ingresar_dataset = input("¿Desea ingresar un nuevo registro a un dataset procesado? (s/n): ")
if ingresar_dataset.lower() not in ['s', 'n']:
    print("Opción no válida. Por favor, ingrese 's' para sí o 'n' para no.")
while ingresar_dataset.lower() == 's':
    dataset_name = input("Ingrese el nombre del dataset al que desea agregar el registro A) iadiza, B) inaturalist, C) xeno_canto): ")
    if dataset_name.upper() not in ['A', 'B', 'C']:
        print("Nombre de dataset no válido. Por favor, ingrese uno de los siguientes: A) iadiza, B) inaturalist, C) xeno_canto.")
    else:
        dataset_index = ord(dataset_name.upper()) - ord('A')  # Convertir A, B, C a índices 0, 1, 2
        header = manejo_archivos.get_header(datasets_paths[dataset_index], **configs[dataset_index])
        nuevo_registro = ejercicio4.generar_registro_vacio(header)
        accion_completar = input("¿Desea completar el registro con información aleatoria? (s/n): ")
        
        if accion_completar == 's' or accion_completar == 'S':
            ejercicio4.completar_registro_random(nuevo_registro)
        else:
            ejercicio4.completar_registro_manual(nuevo_registro)
    
        if ejercicio4.validar_registro(nuevo_registro, **configs[dataset_index]):
            archivo = manejo_archivos.get_archive(datasets_paths[dataset_index], **configs[dataset_index])
            
            if dataset_name.upper() == 'A':
                nuevo_registro['gbifID'] = ejercicio4.generar_id(archivo, dataset_name.upper())
            else:
                nuevo_registro['id'] = ejercicio4.generar_id(archivo, dataset_name.upper())

            ejercicio4.insertar_registro(DATASETS_PATH_PROCESSED, nuevo_registro, archivo , **configs[dataset_index])
            print(f"Registro insertado exitosamente en el dataset {dataset_name.upper()}")

        else:
            print("El registro no es válido y no se ha insertado en el dataset.")
        

    # Única pregunta al final del bucle
    ingresar_dataset = input("\n¿Desea ingresar otro registro? (s/n): ")
    while ingresar_dataset.lower() not in ['s', 'n']:
        ingresar_dataset = input("Por favor, ingrese 's' o 'n': ")
```
