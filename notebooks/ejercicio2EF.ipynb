```python
from utils.constantes import DATASETS_PATH_RAW
from ejercicios.ejercicio2EF import columnas_con_nulos, porcentaje_nulos_por_columna

datasets_paths = [
    DATASETS_PATH_RAW / 'iadiza',
    DATASETS_PATH_RAW / 'inaturalist',
    DATASETS_PATH_RAW / 'xeno_canto'
]
```

```python
resultado = columnas_con_nulos(datasets_paths)

for dataset, columnas in resultado.items():
    print(f"\n{dataset} — {len(columnas)} columnas con nulos:")
    for col in columnas:
        print(f"  - {col}")
```

```python
resultado = porcentaje_nulos_por_columna(datasets_paths)

for dataset, columnas in resultado.items():
    print(f"\n{dataset}:")
    for col, pct in sorted(columnas.items(), key=lambda x: x[1], reverse=True):
        print(f"  {col:<35} {pct:>6.2f}%  ")
```
